# Craigslist Housing Scraper

I designed this program to aid in finding housing in the Boston area and to help myself get back up to speed with coding projects in anticipation of my full-time coding job.

```python
import requests
from bs4 import BeautifulSoup
import re

URL = r"https://boston.craigslist.org/search/boston-ma/apa?lat=42.356&lon=-71.0886&max_price=3100&query=%22august%22%20%7C%20%22aug%22%20%7C%20%228%2F1%22&search_distance=2&sort=date#search=1~gallery~0~0"
```

```python
class Property:
    expression = r"(?<=\/)(\d*?)(?=\.)"
    
    def __init__(self, result) -> None:
        self.title = result.find("div", class_="title").text.strip()
        self.price = result.find("div", class_="price").text.strip()
        self.location = result.find("div", class_="location").text.strip()
        self.URL = result.find("a")['href']
        
        match = re.search(self.expression, self.URL)

        if match:
            self.id = match.group()
        else:
            self.id = 0
    
    def __repr__(self):
        return f'{self.title}\n{self.price}\n{self.location}\n{self.URL}\n{self.id}'
```

```python
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
results = soup.find_all("li", class_="cl-static-search-result")
```
